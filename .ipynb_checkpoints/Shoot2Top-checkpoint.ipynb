```python
import psycopg2
import seaborn as sns
import pandas as pd
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = psycopg2.connect(database="tcount", user="postgres", password="postgres", host="localhost", port="5432")
cur = conn.cursor()
cur.execute("SELECT Urls, Followers from tweets_large")
records = cur.fetchall()
url_list, follower_list = [], []
for rec in records:
  url_list.append(rec[0])
  follower_list.append(rec[1])
conn.commit()
conn.close()
```

```python
tuple_list=zip(url_list, follower_list)
```

```python
tuple_set = set(tuple_list)
```

```python
tweet_count = []
```

```python
tuple_set = list(set(tuple_list))
```

```python
for i in range(len(tuple_set)):
    counter = tuple_list.count(tuple_set[i])
    tweet_count.append(counter)
data = []
for i in range(len(tuple_set)):
    row = tuple_set[i] + (tweet_count[i],)
    data.append(row)
```

```python
data = sorted(data,key=lambda item: item[1], reverse=True )
```

```python
data = pd.DataFrame(data, columns=['Publisher','Followers', 'Tweets'])
```

```python
sns.set(style="whitegrid", color_codes=True)
```

```python
dims = (16, 32)
fig, ax = plt.subplots(figsize=dims)
sns.stripplot(x="Followers", y="Publisher", hue='Tweets', data=data, jitter=True, size=10,ax=ax)
plt.xticks([0,0.05, 0.1, 0.15,0.20,0.25])
plt.plot()
```
